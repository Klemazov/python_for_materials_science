[Функция Вейерштрассе](https://ru.wikipedia.org/wiki/%D0%A4%D1%83%D0%BD%D0%BA%D1%86%D0%B8%D1%8F_%D0%92%D0%B5%D0%B9%D0%B5%D1%80%D1%88%D1%82%D1%80%D0%B0%D1%81%D1%81%D0%B0) построить дома

$$ w(x) = \Sigma_{n=0}^{\infty}b^n\cos(a^n\pi x)$$


where a is an odd number more than 1


and 0<b<1.

```python
# importing the required modules
import numpy as np
import matplotlib.pyplot as plt

# setting the ranges for calculation
# MyVar: the higher the better detalization
MyVar = 100000

# bee for x axis. Here you can change start and stop points
startbee=-2
stopbee=2
bee = np.arange(startbee,stopbee,((stopbee-startbee)/MyVar))

# defining the function. mind indentation! here a=3 and b=1/2
def weierstrass(x,Nvar):
    we=np.zeros(MyVar)
    for n in range(0,Nvar):
        we=we+np.cos(3**n*np.pi*x)/2**n
    return we

# when drawing a plot, we should make x and weierstrass ranges of same
# dimensions
# play with Nvar meaning. my python works properly at 500 and 600
# and fails at 1000 (overflow)
plt.plot(bee,np.reshape(weierstrass(bee, 500),(MyVar,)))

# make the range narrow and enjoy the fractal structure of your plot
plt.show()
```
